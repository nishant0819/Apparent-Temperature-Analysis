# Apparent Temperature Report

## Introduction to dataset

**This dataset is being used here in regarding the analysis of apparent temperature possesed by human body.** Apparent temperature is the temperature that it feels like to our body because of other weather variables other than temperature. This is generally higher in case of high humidity. This is the temperature equivalent perceived by humans, caused by the combined effects of air temperature, relative humidity and wind speed. The measure is most commonly applied to the perceived outdoor temperature. However it also applies to indoor temperatures, especially saunas and when houses and workplaces are not sufficiently heated or cooled.

As Formatted date time is not the column which can be analysed directly so by removing the last string we can make it into the format of date and then new 3 colums can be created through that, i.e. ***month***, ***year*** and ***hour***.

```python
from datetime import datetime
df['Formatted Date']=pd.to_datetime(df['Formatted Date'],utc=True)
df['Formatted Date'] = df['Formatted Date'].astype(str).str[:-6]
```

```python
df['month'] = m.dt.month
```

```python
df['year'] = m.dt.year
```

```python
df['hour'] = m.dt.hour
```

# Bivariate Analysis

We will be taking the above pairs to do bivariate analysis

So, let's start.....

#### Humidity - Temperature

```python
df.plot.scatter(x='Humidity',y='Temperature (C)',color='indigo',animated=True,figsize=(8,8),alpha=0.1)
```

```python
df[['Humidity','Temperature (C)']].describe()
```

We notice that as humidity increases at slow rate temperature decreases at the greater rate that of humidity.


#### Humidity - Apparent Temperature

```python
df.plot.scatter(x='Humidity',y='Apparent Temperature (C)',color='Green',animated=True,figsize=(8,8),alpha=0.1)
```

```python
df[['Apparent Temperature (C)','Humidity']].describe()
```

Apparent temperature shows a very similar change with humidity as that of Temperature showed with it, which lead us to go for the plot of apparent temperature and temperature.



#### Apparent Temperature - Temperature

```python
df.plot.scatter(x='Apparent Temperature (C)',y='Temperature (C)',color='violet', figsize=(8,8),alpha=0.4)
```

```python
df[['Apparent Temperature (C)','Temperature (C)']].describe()
```

We see that there's a linear increase in Temperature as the Apparent Temperature increases. This shows that the properties which make changes in temperature variable can also make similar changes in apparent temperature.



#### Wind Speed - Humidity

```python
df.plot.scatter(y='Humidity',x='Wind Speed (km/h)',color='Black', figsize=(8,8),alpha=0.1)
```

```python
df[['Humidity','Wind Speed (km/h)']].describe()
```

As obvoius, the graph shows that humidity tends to decrease as the wind speed increases.



#### Wind Speed - Temperature

```python
df.plot.scatter(x='Wind Speed (km/h)',y='Temperature (C)',figsize=(8,8),alpha=0.4)
```

Clearly we notice that the increase in wind speed decreases temperature at a proportionate rate.



#### **So  as we can see that increase in wind speed lowers the humidity and as the humidity decreases temperature also goes down, lowering the temperature also lower down the apparent temperature because of their linear relationship.**



#### Month - Apparent Temperature

```python
df.plot.scatter(x='month',y='Apparent Temperature (C)',color='orange', figsize=(8,8),alpha=0.1)
```

```python
df[['Apparent Temperature (C)','month']].describe()
```

It is obvious that as the month increases temperature keeps on increasing, but this led us think more about the year, month and even hourly bases of the change in temperature.

so lets go for detailed datetime.

```python
df.groupby(['year','month'])['Apparent Temperature (C)'].count()
```

This shows that the **data is not uniformly collected**, as in 2005 december we have only 1 single data entry, 2006 november also we have a single entry.

And in other months the data is normally collected.



#### Hour - Apparent Temperature

```python
df.groupby('hour')['Apparent Temperature (C)'].mean().to_frame().plot(kind='bar')
sns.despine()
```

This clearly shows the change in Apparent temperature throughout the day in hourly bases.
